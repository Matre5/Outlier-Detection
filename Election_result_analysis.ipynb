```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.distance import geodesic
```

```python
geolocator = Nominatim(user_agent="my_geocoder", timeout=10)
```

```python
data['Full Address'] = data['State'] + ' ' + data['LGA']
```

```python
def geocode(Full_Address):
    try:
        location = geolocator.geocode(Full_Address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return geocode(Full_Address)
```

```python
data['Coordinates'] = data['Full Address'].apply(geocode)
data[['Latitude', 'Longitude']] = pd.DataFrame(data['Coordinates'].tolist(), index=data.index)
```

```python
radius = 2
```

```python
def neighbouring_cities(row, data, radius = radius):
    neighbors = []
    for index, neighbor in data.iterrows():
        distance = geodesic((row['Latitude'], row['Longitude']), (neighbor['Latitude'], neighbor['Longitude'])).km
        if distance <= radius and index != row.name:
            neighbors.append(index)
    return neighbors
```

```python
FCT_Data['Neighbors'] = data.apply(lambda row: neighbouring_cities(row, data), axis=1)
```

```python
FCT_Data
```

```python
from scipy.stats import zscore
```

```python
def calculate_outlier_scores(data, parties, radius):
    outlier_scores = []
    for index, row in data.iterrows():
        neighbors = neighbouring_cities(row, data, radius)
        scores = {}
        for party in parties:
            if neighbors:
                neighbor_votes = data.loc[neighbors, party]
                mean_votes = neighbor_votes.mean()
                std_dev_votes = neighbor_votes.std()
                if std_dev_votes != 0:
                    scores[party] = (row[party] - mean_votes) / std_dev_votes
                else:
                    scores[party] = 0
            else:
                scores[party] = 0
        scores['Polling Unit'] = row['PU-Name']
        outlier_scores.append(scores)
    return outlier_scores
```

```python
outlier_scores = calculate_outlier_scores(FCT_Data, parties, radius=radius)
outlier_scores_data = pd.DataFrame(outlier_scores)
```

```python
outlier_scores_data
```

```python
FCT_Data['APC_Outlier'] = outlier_scores_data['APC']
```

```python
FCT_Data['LP_Outlier'] = outlier_scores_data['LP']
FCT_Data['PDP_Outlier'] = outlier_scores_data['PDP']
FCT_Data['NNPP_Outlier'] = outlier_scores_data['NNPP']
```

```python
for party in parties:
    sorted_data = FCT_Data.sort_values(by=f'{party}_Outlier', ascending=False)
    print(f"\n Top 3 outliers for {party}: \n")
    print(sorted_data[['PU-Name', 'LGA', f'{party}_Outlier', 'Neighbors']].head(3))
```
